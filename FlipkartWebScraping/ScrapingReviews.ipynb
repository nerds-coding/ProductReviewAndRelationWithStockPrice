```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
```

```python
cols = ['User Rating','User Review','User Name',"Review Date"]
cols2 = ['User Review','User Name',"Review Date"]

all_review_data = pd.DataFrame(columns=cols,index=[0])

all_quality_data = pd.DataFrame(columns=cols2,index=[0])

all_price_value_data = pd.DataFrame(columns=cols2,index=[0])

product_detail_data = pd.DataFrame(columns=['Quality','Details'],index=[0])
```

```python
class ScrapingClass:

    def __init__(self, url):

        self.driver = webdriver.Firefox()
        time.sleep(4)
        self.driver.maximize_window()
        self.driver.get(url)
        time.sleep(6)
        print(self.driver.title)
    
    def writingAllReviewInDataFrame(self,page_source,dataFrame):
        review_div = page_source.find_all('div',{'class':'_3gijNv col-12-12'})
        for j in review_div:
            user_rating = j.find_all('div',{'class':'hGSR34 E_uFuv'})
            user_review = j.find_all('div',{'class':'qwjRop'})

            user_data = j.find_all('div',{'class','row _2pclJg'})
            user_name = [name.find_all('p')[0].text for name in user_data]

            review_data = [name.find_all('p')[-1].text for name in user_data]
            for rating in zip(user_rating,user_review,user_name,review_data):
                print(rating[0].text)
                print(rating[1].text)
                print(rating[2])
                print(rating[3])
                dataFrame.loc[len(dataFrame)] = [rating[0].text, rating[1].text.replace('READ MORE',''), rating[2], rating[3]]
                print()
        print(dataFrame)
    
    def writingReviewInQuality(self,page_source,dataFrame):
        review_div = page_source.find_all('div',{'class':'_3gijNv col-12-12'})
        for j in review_div:
            user_review = j.find_all('div',{'class':'qwjRop'})
            user_data = j.find_all('div',{'class','row _2pclJg'})
            user_name = [name.find_all('p')[0].text for name in user_data]
            review_data = [name.find_all('p')[-1].text for name in user_data]
            
            for rating in zip(user_review,user_name,review_data):
                print(rating[0].text)
                print(rating[1])
                print(rating[2])
                dataFrame.loc[len(dataFrame)] = [rating[0].text.replace('READ MORE',''), rating[1], rating[2]]
                print()
        print(dataFrame.head(10))

    def takingProductSpecs(self):
        time.sleep(10)
        self.driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(10)

        try:
            self.driver.find_element_by_css_selector(".uSQV49").click()
            time.sleep(10)
        except(Exception)as e:
            print(e)
        else:
            productPage = BeautifulSoup(self.driver.page_source, 'html.parser')

        return productPage.find_all('div', {'class': '_2RngUh'})
    
    def readReview(self):
        time.sleep(5)
        total_value = self.driver.find_element_by_css_selector(".swINJg")
        total_reviews = total_value.get_attribute('textContent')
        #clicking on show all reviews
        time.sleep(7)
        self.driver.find_element_by_css_selector(".swINJg > span:nth-child(1)").click()
        time.sleep(3)
        
        # now scrolling on review page to find the number of reviews
        self.driver.execute_script('window.scrollTo(0,document.body.scrollHeight);')
        #taking total number of review pages
        pages = self.driver.find_element_by_css_selector('._2zg3yZ > span:nth-child(1)').text
        
        print(int(pages.split('of')[-1].replace(',','')))
        time.sleep(6)
        self.driver.find_element_by_css_selector('._3fVaIS').click()
        time.sleep(4)

        #review page_source code
        page_code = BeautifulSoup(self.driver.page_source, 'html.parser')
        ScrapingClass.writingAllReviewInDataFrame(self,page_code,all_review_data)
        
        if(',' in pages.split('of')[-1]):
            qua = int(pages.split('of')[-1].replace(',',''))
        else:
            qua = int(pages.split('of')[-1])
 
        for x in range(1542 ,qua):
            #clicking on next button
            time.sleep(6)
            try:
                self.driver.find_element_by_css_selector('a._3fVaIS:nth-child(12)').click()
            except(Exception)as e:
                print(e,type(e))
                try:
                    self.driver.find_element_by_css_selector('a._3fVaIS:nth-child(5)').click()
                except(Exception)as e:
                    print(e,type(e))
                    try:
                        self.driver.find_element_by_css_selector('a._3fVaIS:nth-child(11)').click()
                    except(Exception)as e:
                        try:
                            self.driver.find_element_by_css_selector('a._3fVaIS:nth-child(10)').click()
                        except(Exception)as e:
                            try:
                                self.driver.find_element_by_css_selector('a._3fVaIS:nth-child(9)').click()
                            except(Exception)as e:
                                self.driver.find_element_by_css_selector('a._3fVaIS:nth-child(8)').click()
            finally:
                page_code = BeautifulSoup(self.driver.page_source, 'html.parser')
                ScrapingClass.writingAllReviewInDataFrame(self,page_code,all_review_data)   
                time.sleep(4)

            #review page_source code
#             page_code = BeautifulSoup(self.driver.page_source, 'html.parser')
#             ScrapingClass.writingAllReviewInDataFrame(self,page_code,all_review_data)
  
        return page_code
    
    def readQualityReview(self):
#         time.sleep(5)
#         total_value = self.driver.find_element_by_css_selector(".swINJg")
#         total_reviews = total_value.get_attribute('textContent')
#         #clicking on show all reviews
#         time.sleep(7)
#         self.driver.find_element_by_css_selector(".swINJg > span:nth-child(1)").click()
        time.sleep(3)
        
        self.driver.find_element_by_css_selector('._36Dmoj > div:nth-child(1) > a:nth-child(2)').click()
        time.sleep(5)
        
        # now scrolling on review page to find the number of reviews
        self.driver.execute_script('window.scrollTo(0,document.body.scrollHeight);')
        #taking total number of review pages
        pages = self.driver.find_element_by_css_selector('._2zg3yZ > span:nth-child(1)').text
        
        print(int(pages.split('of')[-1]))
        time.sleep(6)
        self.driver.find_element_by_css_selector('._3fVaIS').click()
        time.sleep(4)

        #review page_source code
        page_code = BeautifulSoup(self.driver.page_source, 'html.parser')
        ScrapingClass.writingReviewInQuality(self,page_code,all_quality_data)
        
        if(',' in pages.split('of')[-1]):
            qua = int(pages.split('of')[-1].replace(',',''))
        else:
            qua = int(pages.split('of')[-1])
        
        for x in range(int(pages.split('of')[-1])):
            #clicking on next button
            time.sleep(5)
            #self.driver.find_element_by_css_selector('a._3fVaIS:nth-child(12)').click()
            try:
                self.driver.find_element_by_css_selector('a._3fVaIS:nth-child(12)').click()
            except(Exception)as e:
                print(e,type(e))
                try:
                    self.driver.find_element_by_css_selector('a._3fVaIS:nth-child(5)').click()
                except(Exception)as e:
                    print(e,type(e))
                    try:
                        self.driver.find_element_by_css_selector('a._3fVaIS:nth-child(11)').click()
                    except(Exception)as e:
                        try:
                            self.driver.find_element_by_css_selector('a._3fVaIS:nth-child(10)').click()
                        except(Exception)as e:
                            try:
                                self.driver.find_element_by_css_selector('a._3fVaIS:nth-child(9)').click()
                            except(Exception)as e:
                                self.driver.find_element_by_css_selector('a._3fVaIS:nth-child(8)').click() 
            finally:
                #review page_source code
                page_code = BeautifulSoup(self.driver.page_source, 'html.parser')
                ScrapingClass.writingReviewInQuality(self,page_code,all_quality_data)
                time.sleep(4)
        
        self.driver.close()
        return page_code
    
```
